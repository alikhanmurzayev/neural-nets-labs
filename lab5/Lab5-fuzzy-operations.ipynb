* Division in the method __truediv__

```python
import numpy as np
import matplotlib.pyplot as plt
from functools import partial
from collections import defaultdict
```

```python
def mu_max(val, width, x):
    buff = ((x - val) / width) * ((x - val) / width)
    return max(0, 1-buff)  


class FuzzyValue_V2():
    
    def __init__(self, val, width, mu=None):
        self.val = float(val)
        self.width = float(width)
        if mu is not None:
            self.mu = partial(mu, self.val, self.width)
        self.op = None

    def __call__(self, x):
        if self.mu is not None:
            return float(self.mu(x))

    def __truediv__(self, o):
        self.op = o
        
        left = (self.val - self.width) / (self.op.val + self.op.width)
        right = (self.val + self.width) / (self.op.val - self.op.width)
        
        val = (right + left) / 2
        width = (right - left) / 2
        
        return FuzzyValue_V2(val, width, mu=mu_max)
    
    
    def draw(self):
        xs = np.linspace(-7, 20, num=200)
        y = [self(x) for x in xs]
        plt.figure()
        plt.plot(xs, y)        
        
    def __add__(self, o): 
        self.op = o
        
        left = (self.val - self.width) + (self.op.val - self.op.width)
        right = (self.val + self.width) + (self.op.val + self.op.width)
                
        val = (right + left) / 2
        width = (right - left) / 2
        
        return FuzzyValue_V2(val, width, mu=mu_max)

    def __sub__(self, o):
        self.op = o
        
        left = (self.val - self.width) - (self.op.val + self.op.width)
        right = (self.val + self.width) - (self.op.val - self.op.width)
        
        val = (right + left) / 2
        width = (right - left) / 2
        
        return FuzzyValue_V2(val, width, mu=mu_max)
        

a = FuzzyValue_V2(5.1, 1.5, mu_max)
b = FuzzyValue_V2(4.3, 2, mu_max)
```
